```python
applicants=pd.read_excel("Applicants.xlsx")
tasks=pd.read_excel("Tasks.xlsx")
```

```python
x=applicants[["x_coord"]].to_numpy()
y=applicants[["y_coord"]].to_numpy()
x_t=tasks[["x_coord"]].to_numpy()
y_t=tasks[["y_coord"]].to_numpy()
```

```python
plt.scatter(x, y, c='green', alpha=0.8,s=1,label="Volunteers")
plt.scatter(x_t,y_t, c="red",alpha=0.8,s=40,label="Tasks",marker="*")

# plt.title('Locations in 2D Plane')
plt.legend(loc='best',fontsize=11)
plt.xlabel('x-axis',)
plt.ylabel('y-axis')
plt.savefig('D:\\python_apps\\MTP\\Plots\\location_dist.eps', format='eps',dpi=300)
plt.show()
```

```python
distances=[]
minm=10000
maxm=0
for i in range (0,97):
    for j in range(0,1575):
        pnt1=np.array((x_t[i],y_t[i]))
        pnt2=np.array((x[j],y[j]))
        dist=np.linalg.norm(pnt1-pnt2)
        maxm = max(maxm, dist)
        minm = min(minm, dist)
        distances.append(dist)
   
```

```python
avg_dist=np.average(distances)  
```
